```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime

# Data set https://www.kaggle.com/berkeleyearth/climate-change-earth-surface-temperature-data
cty_temp = pd.read_csv('GlobalLandTemperaturesByCountry.csv')
```

```python
# Some features to the data needed for analysis
cty_temp['dt']         = pd.to_datetime(cty_temp['dt'])
cty_temp['month']      = cty_temp['dt'].dt.strftime('%m')
cty_temp['year']       = cty_temp['dt'].dt.strftime('%Y')
cty_temp['yearmonth']  = cty_temp['dt'].dt.strftime('%Y%m')
# Adding integer values of month, year and yearmonth to make it easier to graph it.
# TODO Better way to do this?
cty_temp['yeari']      = cty_temp['year'].apply(pd.to_numeric)
cty_temp['monthi']     = cty_temp['month'].apply(pd.to_numeric)
cty_temp['yearmonthi'] = cty_temp['yearmonth'].apply(pd.to_numeric)

# Filter out some data and print out the structure
cty_temp_alnd = cty_temp.loc[lambda df: (cty_temp['Country'] == 'Åland'), :]
cty_temp_alnd[:10]
```

```python
# Select temp for year 1900 and country Aland. An exotic little island in the Baltic sea
cty_temp_aland = cty_temp[ (cty_temp['year'] == '1900') & (cty_temp['Country'] == 'Åland') ]
# Display average temperature in line plot
cty_temp_aland.plot(kind = 'line', x = 'monthi',y = 'AverageTemperature')
```

Average temparature for Aland is as expected. Higher during the summer months (month 5-9) and lower during winter months (month 1-4, 10-12)

```python
# Select temp for year 1900 and country Aland and Sweden.
cty_temp_aland = cty_temp[ (cty_temp['year'] == '1900') & (cty_temp['Country'] == 'Åland') ]['AverageTemperature']
cty_temp_sweden = cty_temp[ (cty_temp['year'] == '1900') & (cty_temp['Country'] == 'Sweden') ]['AverageTemperature']

# Create subplot for each country and display the data
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(cty_temp_aland, color='purple')
plt.ylabel('Avg temp')
plt.title('Temperature - Aland')

plt.subplot(1, 2, 2)
plt.plot(cty_temp_sweden, color='green')
plt.ylabel('Avg temp')
plt.title('Temperature - Sweden')

plt.tight_layout()
plt.show()
```

Above graphs display temperature for Aland and Sweden for year 1900. Very similar as it expected, since they are right next to each other.

```python
# Select temperature in october and Sweden for all years we have measurements
cty_temp_sweden_oct = cty_temp[ (cty_temp['month'] == '10') & (cty_temp['Country'] == 'Sweden') ]
# Display a scatter plot of year and temperature
cty_temp_sweden_oct.plot(kind='scatter', x='yeari', y='AverageTemperature')
```

The above scatter graphs hints on weak rising curve. Could be added using `np.polyfit()`

```python
# get mean for temperature grouped by year
cty_temp.groupby('yeari').mean()['AverageTemperature'].plot()
```

Above graph indicates that measurements before 1900 is more sparse than after. Assume data set is sparse since there are relatively big fluctuations. Also colder places were measured before 1900. Which could indicate than we have measurements primarily for Nothern Europe, Nothern America. Further investigation into the dataset can validate my assumptions.
